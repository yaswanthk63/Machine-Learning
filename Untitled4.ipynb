# Visualizing CNN Feature Maps: Understanding Deep Learning Representations

This notebook demonstrates how Convolutional Neural Networks (CNNs) learn internal representations by visualizing feature maps at different layers.

The goal is to understand:
- How CNNs extract features
- Differences between early and deep layers
- How feature maps evolve during learning

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
import seaborn as sns

# Reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Clean visual theme
sns.set_theme(style="white")

print("Libraries loaded successfully.")
```

### Load data

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Reshape for CNN
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

print("Dataset loaded successfully.")
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

### Train Model

```python
model.fit(
    X_train, y_train,
    epochs=3,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)
```

```python
img = X_test[0]

plt.figure(figsize=(4,4))
plt.imshow(img.squeeze(), cmap='gray')
plt.title("Input Image", fontsize=12)
plt.axis('off')

plt.savefig("figure1_input.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Feature Model

```python

_ = model.predict(X_test[:1])

# Now safely extract outputs
layer_outputs = [layer.output for layer in model.layers if 'conv' in layer.name]

feature_model = tf.keras.models.Model(inputs=model.inputs, outputs=layer_outputs)

print("Feature extraction model created successfully.")
```

### Feature Maps

```python
feature_maps = feature_model.predict(img.reshape(1,28,28,1))
```

### Layer - 1

```python
maps = feature_maps[0]

plt.figure(figsize=(10,6))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(maps[0,:,:,i], cmap='cividis')
    plt.axis('off')

plt.suptitle("Feature Maps - Layer 1", fontsize=14)
plt.tight_layout()

plt.savefig("figure2_layer1.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Layer - 2

```python
maps = feature_maps[1]

plt.figure(figsize=(10,6))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(maps[0,:,:,i], cmap='cividis')
    plt.axis('off')

plt.suptitle("Feature Maps - Layer 2", fontsize=14)
plt.tight_layout()

plt.savefig("figure3_layer2.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Deep Layer

```python
maps = feature_maps[2]

plt.figure(figsize=(10,6))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(maps[0,:,:,i], cmap='cividis')
    plt.axis('off')

plt.suptitle("Feature Maps - Deep Layer", fontsize=14)
plt.tight_layout()

plt.savefig("figure4_deep_layer.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Grid Visualization

```python
layer_index = 0  # Change to 1 or 2

feature_map = feature_maps[layer_index]
num_filters = feature_map.shape[-1]

cols = 8
rows = int(np.ceil(num_filters / cols))

plt.figure(figsize=(12, rows * 2))

for i in range(num_filters):
    plt.subplot(rows, cols, i+1)
    plt.imshow(feature_map[0, :, :, i], cmap='cividis')
    plt.axis('off')

plt.suptitle(f"Grid Feature Maps - Layer {layer_index+1}", fontsize=14)
plt.tight_layout()

plt.savefig(f"figure5_grid_layer{layer_index+1}.png", dpi=300, bbox_inches='tight')
plt.show()
```
